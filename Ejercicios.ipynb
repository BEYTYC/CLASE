## 1. 🎯 Ejemplo de un tirador

**Problema**:  
*Un tirador quiere acertar dentro de un círculo de radio $r = 3$ en un cuadrado de lado $s = 12$.  
¿Cuál es la probabilidad de que al disparar, acierte dentro del círculo?*


```python
import matplotlib.pyplot as plt
import numpy as np

# Parámetros
s = 12
r = 3

# Cálculo de áreas
area_cuadrado = s ** 2
area_circulo = np.pi * r ** 2
prob_acierto = round(area_circulo / area_cuadrado, 2)

# Visualización
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.set_xlim(0, s)
ax.set_ylim(0, s)

# Cuadrado
square = plt.Rectangle((0, 0), s, s, edgecolor='black', facecolor='none', linewidth=2)
ax.add_patch(square)

# Círculo
circle = plt.Circle((s/2, s/2), r, edgecolor='red', facecolor='none', linewidth=2)
ax.add_patch(circle)

# Textos informativos
ax.text(s / 2, s * 0.9, f"Área Cuadrado = {area_cuadrado}", ha='center', fontsize=10)
ax.text(s / 2, s * 0.1, f"Área Círculo = {round(area_circulo, 2)}", ha='center', fontsize=10, color='red')
ax.set_title("Probabilidad Geométrica: Punto en un Área")

plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()

# Mostrar la probabilidad
print(f"P(A) = {prob_acierto}")
```

## 2. 📌 Problema: Encuentro aleatorio entre dos amigos

Dos amigos acuerdan encontrarse entre las **12:00 PM y la 1:00 PM**.  
Cada uno llega en un momento aleatorio e independiente durante esa hora.  
El primero que llega espera 15 minutos, luego se va.  

### ¿Cuál es la probabilidad de que se encuentren?

---

### ✏️ Modelando el problema

Sean:

- $X$: tiempo de llegada del primer amigo  
- $Y$: tiempo de llegada del segundo amigo  

Ambos valores son aleatorios y uniformemente distribuidos entre $[0, 60]$.  
Se encuentran si:

$$
|X - Y| \leq 15
$$

---

### 📐 Representación geométrica

Todos los pares $(X, Y)$ posibles forman un cuadrado de $60 \times 60$.  
La región de encuentro está limitada por:

$$
Y - 15 \leq X \leq Y + 15
$$

Lo que genera una banda de ancho 30 minutos centrada en la diagonal $X = Y$.

---

### 🔢 Cálculo del área

- Área total: $A_{\text{total}} = 60 \times 60 = 3600$
- Área no favorable (dos triángulos): $2 \times \frac{1}{2} \times 45 \times 45 = 2025$
- Área favorable: $1575$

$$
P = \frac{1575}{3600} = \frac{7}{16} \approx 0.4375
$$


```python
# Visualización del problema de los amigos

# Parámetros
t_max = 60
espera = 15

# Crear figura
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_aspect('equal')
ax.set_xlim(0, t_max)
ax.set_ylim(0, t_max)

# Cuadro general
ax.add_patch(plt.Rectangle((0, 0), t_max, t_max, edgecolor='gray', facecolor='none', linewidth=2))

# Región de encuentro (banda)
x_poly = [0, t_max - espera, t_max, espera]
y_poly = [espera, t_max, t_max - espera, 0]
ax.fill(x_poly, y_poly, color='forestgreen', alpha=0.3)

# Límites superior/inferior
ax.plot([0, t_max], [espera, t_max + espera], linestyle='--', color='darkgreen', linewidth=2)
ax.plot([0, t_max], [-espera, t_max - espera], linestyle='--', color='darkgreen', linewidth=2)

# Diagonal
ax.plot([0, t_max], [0, t_max], linestyle=':', color='blue', linewidth=2)

# Etiquetas y leyenda
ax.set_xlabel("Llegada del Amigo 1 (min)")
ax.set_ylabel("Llegada del Amigo 2 (min)")
ax.set_title("📐 Región donde los amigos se encuentran")

from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], color='forestgreen', lw=6, alpha=0.3, label='Región de encuentro'),
    Line2D([0], [0], color='darkgreen', lw=2, linestyle='--', label='Límites de espera'),
    Line2D([0], [0], color='blue', lw=2, linestyle=':', label='Llegan al mismo tiempo')
]
ax.legend(handles=legend_elements, loc='upper right')

plt.grid(True)
plt.show()
```
